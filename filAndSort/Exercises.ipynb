### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")
df
```

### Step 3. Assign it to a variable called chipo.

```python
cols = df.columns
for col in cols:
    print(df[col].dtype)
```

```python
str = "$9.36"
splt = lambda s: s[1:]
ans = splt(str)
print(ans)
```

```python
df["item_price"] = df["item_price"].apply(splt)
df
```

```python
df["item_price"] = df["item_price"].astype(float)
cols = df.columns
for col in cols:
    print(df[col].dtype)
```

### Step 4. How many products cost more than $10.00?

```python
df_prod = df[df["item_price"] > 10]
df_prod
```

```python
counter=0
for i in df_prod["item_name"]:
    counter+=1

print(counter)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
df_item_price = df[["item_name", "item_price"]]
df_item_price
```

### Step 6. Sort by the name of the item

```python
df = df.sort_values(by=["item_name"])
df
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price = df["item_price"].max()
print(max_price)
```

```python
val = df.loc[df["item_price"] == max_price]["quantity"]
print(val)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
count_items = df.value_counts(["item_name"]).reset_index()
count_items
```

```python
count_items[count_items["item_name"] == "Veggie Salad Bowl"]["count"]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
df[(df["item_name"] == "Canned Soda") & (df["quantity"] > 1)].count()

```
